```python
import pandas
import configparser
import psycopg2
```

```python
config= configparser.ConfigParser()
config.read('config.ini')

host= config['myaws']['host']
user=config['myaws']['user']
pwd=config['myaws']['pwd']
database= config['myaws']['database']
```

```python
conn= psycopg2.connect(
                        host= host,
                            user=user,
                            password=pwd,
                            dbname= database)
```

```python
cur=conn.cursor()
```

# Q1

```python
sql_statement=  '''
                select * 
                from student_gp6
                '''
```

```python
df=pandas.read_sql_query(sql_statement,conn)
df[:]
```

# Q2

```python
sql_statement=  '''
                select cname, count(semail) from enroll_gp6 group by cname
                order by count(semail) desc 
                limit 1
                '''
```

```python
df=pandas.read_sql_query(sql_statement,conn)
df[:]
```

# Q3

```python
sql_statement=  '''
                select 
                *
                from teacher_gp6
                inner join course_gp6
                on teacher_gp6.email= course_gp6.temail
                '''
```

```python
df=pandas.read_sql_query(sql_statement
                ,conn)
df[:]
```

# Q4

```python
sql_statement=  '''
                select teacher_gp6.name as tname,
                count(*) as course_num
                from teacher_gp6
                inner join course_gp6
                on teacher_gp6.email=course_gp6.temail
                group by teacher_gp6.name
                order by course_num desc
                limit 1
                '''
```

```python
df=pandas.read_sql_query(sql_statement,conn)
df[:]
```
